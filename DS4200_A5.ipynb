```python
vg_df = pd.read_csv('vgsales.csv')
vg_df
```

```python
# Interactive and Color coded scatter plot which tells us the numebr of 
# sales of the top selling video games in North America
# can also zoom in and out of graph 

vg_5 = vg_df[vg_df['Year'] == 2002.0]
vg_5 = vg_df[vg_df['NA_Sales'] > 5]


title = alt.TitleParams('Sales of the most popular video Games in North America in 2002', anchor='middle')
alt.Chart(vg_5, title=title).mark_circle(size=60).encode(
    x=alt.X('Name',title='Game Titles'),
    y=alt.Y('NA_Sales',title='Sales in North America (millions)'),
    color='Genre',
    tooltip=['Name', 'Platform', 'Genre', 'NA_Sales']
).interactive()
```

```python

brush = alt.selection_interval(encodings=['x'])

base = alt.Chart(vg_5).mark_area().encode(
    x=alt.X('Year:O'),
    y=alt.Y('sum(Global_Sales):Q', title='Total Video Game Sales'),
    tooltip=["sum(Global_Sales)", 'Year'],
)

selected = base.transform_filter(brush).mark_area(color='goldenrod')

chart1 = alt.layer(base.add_selection(brush), selected).properties(title="Total Video Games Sales Every Year")
chart1
```
